```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.shape
df.info()
```

```python
df2=df[['LotArea', 'FullBath', 'HalfBath', 'BedroomAbvGr','SalePrice']]
df2.head()
```

```python
df2.drop_duplicates()
```

```python
#correlation matrix
fig, axs = plt.subplots(figsize=(8, 4))
mat = df2.corr(method = 'pearson')
mask = np.triu(np.ones_like(mat, dtype=bool))
sns.heatmap(mat, mask=mask, cmap = sns.color_palette('Blues'), vmax=1, center=0, annot = False, linewidths=.5, cbar_kws={'shrink': .7})
axs.set_title("Correlation Map")
plt.show()
```

```python
fig, axs = plt.subplots(figsize = (16,6))

axs = plt.scatter(x=df2.LotArea, y=df.SalePrice)
plt.ylabel('Sale Price')
plt.xlabel('Lot size in square feet')
plt.title('Scatter Relating SalePrice to LotArea')
```

```python
def detectOutliers():
    fig, axs = plt.subplots(2,3, figsize = (10,5))
    plt1 = sns.boxplot(df2['SalePrice'], ax = axs[0,0])
    plt2 = sns.boxplot(df2['LotArea'], ax = axs[0,1])
    plt3 = sns.boxplot(df2['FullBath'], ax = axs[0,2])
    plt1 = sns.boxplot(df2['HalfBath'], ax = axs[1,0])
    plt2 = sns.boxplot(df2['BedroomAbvGr'], ax = axs[1,1])
    plt.tight_layout()
detectOutliers()
```

```python
# Outlier reduction for price
Q1 = df2.SalePrice.quantile(0.25)
Q3 = df2.SalePrice.quantile(0.75)
IQR = Q3 - Q1
df2 = df2[(df2.SalePrice >= Q1 - 1.5*IQR) & (df2.SalePrice <= Q3 + 1.5*IQR)]
# Outlier reduction for area
Q1 = df2.LotArea.quantile(0.25)
Q3 = df2.LotArea.quantile(0.75)
IQR = Q3 - Q1
df2 = df2[(df2.LotArea >= Q1 - 1.5*IQR) & (df2.LotArea <= Q3 + 1.5*IQR)]

# Outlier reduction for area
Q1 = df2.BedroomAbvGr.quantile(0.25)
Q3 = df2.BedroomAbvGr.quantile(0.75)
IQR = Q3 - Q1
df2 = df2[(df2.BedroomAbvGr >= Q1 - 1.5*IQR) & (df2.BedroomAbvGr <= Q3 + 1.5*IQR)]
detectOutliers()
```

```python
df2.dtypes
```

```python
X = df2.drop('SalePrice',axis=1)
y = df2['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=33)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
r2 = round(r2_score(y_test,y_pred),5)
print('Coefficient of determination R2: ', r2)
```

```python
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='-')
plt.xlabel('Actual Values')
plt.ylabel('Prediction')
plt.title('Scatter chart - Linear regression model')
```
